```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ConsumoCo2.csv")
```

# Divisão de treino e teste

```python
from sklearn.model_selection import train_test_split

X = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB']]
y = df['CO2EMISSIONS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Árvore de Decisão

```python
from sklearn.tree import DecisionTreeRegressor

ad_model = DecisionTreeRegressor()
ad_model.fit(X_train, y_train)
```

- Esse algoritmo treina o modelo de Árvore de Decisão.
- Durante o treinamento, o algoritmo aprende a construir uma árvore de decisão que mapeia as features nas variáveis previsoras X_train para os valores da variável de destino y_train

# Navies Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
```

- Esse algoritmo treina o modelo Naive Bayes.
- Durante o treinamento, o algoritmo aprende a estimar as probabilidades de um dado pertencer a uma classe (valor de emissão de CO2) baseado nas features nas variáveis previsoras X_train

# Avaliação de Modelos

```python
from sklearn.metrics import mean_squared_error


ad_mse = mean_squared_error(y_test, ad_model.predict(X_test))
nb_mse = mean_squared_error(y_test, nb_model.predict(X_test))

print('Erro Médio Quadrático (MSE) - Árvore de Decisão:', ad_mse)
print('Erro Médio Quadrático (MSE) - Naive Bayes:', nb_mse)
```
